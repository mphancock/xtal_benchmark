```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

import os 
import pickle
```

```python
file = "/Users/matthew/Documents/phd/xtal/benchmark/data/score_vs_rmsd_cluster.p"
with open(file, 'rb') as f: 
    data = pickle.load(f)
    
file_nmr = "/Users/matthew/Documents/phd/xtal/benchmark/data/1ubq_nmr_score_vs_rmsd.p"
with open(file_nmr, 'rb') as f: 
    data_nmr = pickle.load(f)
    
file_md = "/Users/matthew/Downloads/1ubq_md_score_vs_rmsd.p"
with open(file_md, 'rb') as f: 
    data_md = pickle.load(f)
```
